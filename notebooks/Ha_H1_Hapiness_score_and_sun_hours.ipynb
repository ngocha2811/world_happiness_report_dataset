```python
import pandas as pd

url = 'https://raw.githubusercontent.com/ngocha2811/world_happiness_report_dataset/refs/heads/main/data/clean/happiness_weather_data'

df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
df_filtered = df[['monthly_sun_hours', 'country','happiness_rank','happiness_score','economy_(gdp_per_capita)','daily_temp','monthly_rain']]
```

```python
df_filtered
```

```python
df_filtered.sort_values(by='monthly_sun_hours')
```

```python
countries_drop = ['Russia','Kazakhstan']
```

```python
df_filtered = df_filtered[~df_filtered["country"].isin(countries_drop)]
```

```python
df_filtered.sort_values(by='monthly_sun_hours')
```

```python
df_filtered.dropna(subset='monthly_sun_hours',inplace=True)
```

```python
df_filtered
```

```python
df= df_filtered
```

```python
df['sun_happiness'] = df['monthly_sun_hours']/df['happiness_score']
```

```python
df
```

```python
df.plot.scatter(x='monthly_sun_hours', y='happiness_score')
```

```python
df['monthly_sun_hours'].corr(df['happiness_score'])
```

```python
import numpy as np
import matplotlib.pyplot as plt 
m, b = np.polyfit(df['monthly_sun_hours'], df['happiness_score'], 1)

plt.scatter(df['monthly_sun_hours'], df['happiness_score'])
plt.plot(df['monthly_sun_hours'], m*df['monthly_sun_hours'] + b)
plt.xlabel("monthly_sun_hours")
plt.ylabel("happiness_score")
plt.show()
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ngocha2811/world_happiness_report_dataset/refs/heads/main/world_happiness_report.csv"

df = pd.read_csv(url)
regions_to_keep = ["Western Europe", "Central and Eastern Europe"]
df_filtered = df[df["Region"].isin(regions_to_keep)]
df_filtered.reset_index(drop=True)
df_filtered = df_filtered.drop(columns = ["Unnamed: 0"])
df_filtered
```

```python
def cleaning_df(df):
    df.columns = [col.lower().replace(" ", "_") for col in df.columns]
    print(f"Dataset has {df.duplicated().sum()} duplicated values")
    return df
```

```python
df_filtered = cleaning_df(df_filtered)
df_filtered = df_filtered[df_filtered["year"] == 2015]
```

```python
url2015 ='https://raw.githubusercontent.com/ngocha2811/world_happiness_report_dataset/refs/heads/main/data/raw/weather_2015.csv'
url2016= 'https://raw.githubusercontent.com/ngocha2811/world_happiness_report_dataset/refs/heads/main/data/raw/weather_2016.csv'


def generate_df_weather_data(url):
    df = pd.read_csv(url)
    df = df.groupby("point_id", as_index=False).mean(numeric_only=True)
    df = df.rename(columns={'point_id' : 'country',
    "tsun": "monthly_sun_hours",
    "tavg": "daily_temp",
    "prcp":"monthly_rain"})
    df = df[['country',"monthly_sun_hours", "daily_temp","monthly_rain" ]]
    df['monthly_sun_hours'] = df['monthly_sun_hours']/60
    return df
```

```python

df2015 = generate_df_weather_data(url2015)
```

```python
df = pd.merge(df_filtered, df2015, on='country', how='left')
```
